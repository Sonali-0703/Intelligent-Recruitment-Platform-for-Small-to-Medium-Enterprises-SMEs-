```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
data = pd.read_csv('candidate_data.csv')

# Histogram of Candidate Skills
plt.figure(figsize=(10, 6))
sns.histplot(data['skills'], bins=20, kde=True)
plt.title('Distribution of Candidate Skills')
plt.xlabel('Skills')
plt.ylabel('Frequency')
plt.show()

# Pie Chart of Job Role Distribution
job_roles = data['job_role'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(job_roles, labels=job_roles.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Roles')
plt.show()

# Match Score Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['match_score'], bins=20, kde=True)
plt.title('Distribution of Match Scores')
plt.xlabel('Match Score')
plt.ylabel('Frequency')
plt.show()
```
